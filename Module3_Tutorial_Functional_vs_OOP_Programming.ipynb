# Sort an Array of os, 1s, and 2s



```python
class Solution:
    def sort012(self, arr):
        # Count the occurences of each unique element (0, 1, and 2)
        # Step 1: Count the numbers
        cnt0 = arr.count(0)
        cnt1 = arr.count(1)
        cnt2 = arr.count(2)
        
        # Step 2: Overwrite the original array
        i = 0
        for _ in range(cnt0):
            arr[i] = 0
            i += 1
        for _ in range(cnt1):
            arr[i] = 1
            i += 1
        for _ in range(cnt2):
            arr[i] = 2
            i += 1
        return arr

if __name__ == "__main__":
    # Create the 'sol' object
    sol = Solution()
```

# Binary Search


```python
# Binary Search

class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        result = -1  # Default if k is not found
        
        while low <= high:
            mid = (low + high) // 2
            
            if arr[mid] == k:
                result = mid      # Record the index
                high = mid - 1    # Look for smaller index
            elif arr[mid] < k:
                low = mid + 1     
            else:
                high = mid - 1    
                
        return result
```
